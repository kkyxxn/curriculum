```python
import random

while 1: # 무한루프
    print("""  
원하는 주제를 고르시면 문제가 출제 됩니다.
1.넌센스
2.수도
3.영화 자음""") # 초기화면 세팅

    a=input("번호 : ") # 입력받은 숫자
    
    if a not in('1','2','3'):  # 1,2,3이 아니면
        print('*올바른 숫자를 입력해주세요.\n\n') # 해당 창 띄우고 
        continue # 다시 숫자 받어
        
        
    # 정상 숫자 입력되면 (1,2,3) 문제1,2,3 읽어와
    with open(f'문제{a}.txt', 'r', encoding="utf-8") as f: 
        questions = f.readlines(); # 문제 readlines 전체 읽어 오기
        random.shuffle(questions) # questions 랜덤 섞기
    
    print('=-='*5,'퀴즈시작','=-='*5) # 퀴즈 시작 화면 

    count = 0 # 문제 개수
    score = 0 # 맞은 개수

    for s in questions: # questions에서 하나씩 뽑아 s에 할당
        count += 1 # 카운팅 1부터
        tmp = s.strip().split('\t') # 양쪽 공백 제거 후, \t(탭)으로 구분하고 tmp에 담기
        
        print(f'{count}번 문제 : {tmp[0]}') # 1번 문제 : tmp[0] 문제
        
        if input('답 : ') == tmp[1]: # 사용자의 답이 tmp[1]과 같으면
            print('◎ 정답입니다!\n');score += 1  # 정답 출력 후 , 맞은 개수 늘리기
        else:
            print(f"X 틀렸습니다! 정답은 '{tmp[1]}' 입니다!\n") # 틀리면 tmp[1]에 들어가있는 정답 값 출력
        
        if count==5:
            break # 그러다 문제 개수가 5개면 break 줌

    print(f'총 맞춘 갯수 : {score}') # quiz 끝났을 때 화면 출력

    print("""
0개 아쉬워요
1~2개 잘했어요 
3~4개 평균이상
5개  전문가
    """)
```
