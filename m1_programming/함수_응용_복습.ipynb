```python
def factorial(n):
    output=1 # 초기값을 1로줌 0으로 하면 곱해봐야 모두 0이므로
    for i in range(1,n+1): # range 함수의 특징때문에 n+1로 설정 --> range함수는 처음부터 끝깞-1이므로
        output *=i 
    return output
print(factorial(5))
print(factorial(10))
```

```python
# 재귀란 '자기자신을 호출하는 것' 의미
# 재귀함수를 이용해서 팩토리얼을 구하는 사용자 함수를 작성
# factorial(n)=  n*factorial(n-1) (n>=1)

def factorial(n):
    if n==0: # 0 일때는 1로 retrun하게 해주고 
        return 1 
    else: # 0이 아닌 수면 원래 factorial 대로 
        return n* factorial(n-1)

print(factorial(0)) # 0 factorial은 1
print(factorial(5))
```

```python
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina', 'West virginia']

def clean_strings(strings): #사용자 함수 정의
    result =[]                # 결과값을 담을 변수와 리스트설정
    for value in strings:        # strings에서 하나씩 꺼내 value에 넣어줌
        value = value.strip()     # 받은 value값을 strip 공백 제거해줌
        result.append(value)          # 결과값에 할당. append
    return result                      #결과값 리턴
clean_strings(states)                   #호출
```

```python
import re #정규표현식 패키지 
states = ['   Alabama ', 'Georgia!', 'Georgia', 'georgia', 'FlOrIda',
          'south   carolina##', 'West virginia?']
def clean_strings(strings):
    result = []
    for value in strings:
        value=value.strip() # 양쪽 공백제거
        value=re.sub('[!#?]','',value) # 특수문자 제거
        value=value.title() # 첫 글자만 대문자로 변경
        result.append(value)
    return result

clean_strings(states)
```

```python
# 단어별 개수 구하기
def getTextFreq(filename):
    with open(filename,'r') as f:
        text = f.read()
        fa={} # key, value 형태로 선언
        for c in text: # text에 있는 각 문자에 대해서
            if c in fa: # fa에 c가 있으면
                fa[c]+=1 # count 하는거니까 count 1 추가해줘야함 
            else: # 포함되어 있지 않으면
                fa[c]=1 # 그냥 1적용
    return fa # return은 with 라인에 들여쓰기
ret= getTextFreq('C:\cakd5\workspace\data\mydata.txt') # 사용자함수 이용해서 ('파일경로 설정')
ret= sorted(ret.items(), key=lambda x:x[1], reverse=True) # 내림차순
print(ret) # 출력
```

```python
# 연도/성별/출생아수를 구하세요. -> 1880, 90994, 110490
for i in range(1880,2016):
    with open(f'data/names/yob{i}.txt','r') as f:
        M_sum=0
        F_sum=0
        for line in f:
            (name,sex,same) = line.strip().split(',')
            if sex =='M':
                M_sum += int(same)
            else:
                F_sum += int(same)
    print(f'{i}년 / 여아 {F_sum}명 / 남아 {M_sum}명')
```

```python
# 연도별 인기있는 상위 10개 성별 출생아 이름 구하기(특정연도 입력하면 해당연도의 출생아 이름 출력)
# TOP_1 여자아기이름:('Emily', 25057)

year=input('연도를 입력하세요 (1880~2019) > ')
filename = 'data/names/yob' + year + '.txt'
print()

with open(filename) as file:
    f_fav=[]
    m_fav=[]
    for line in file:
        (name, gender, birth) = line.strip().split(',')
        if(gender == 'F'):
            f_fav.append((name, int(birth)))
        else:
            m_fav.append((name, int(birth)))
    del f_fav[10:]
    del m_fav[10:]
        
    print(f'{year}년도')
    for i in range(10):
        print(f'\tTOP_{i+1} 여자아이 이름: {f_fav[i][0]} / {f_fav[i][1]}명') #  문자열안에서 줄바꿈 : \n / 탭 : \t
    print('\t'+'='*45)
    for i in range(10):
        print(f'\tTOP_{i+1} 남자아이 이름: {m_fav[i][0]} / {m_fav[i][1]}명')
```

```python
from os.path import exists

def getTop10BabyName(year):
    nameF = {}
    nameM = {}
    
    filename = 'data/names/yob%s.txt'%year
    if not exists(filename):
        print('[%s] 파일이 존재하지 않습니다.'%filename)
        return None
    
    with open(filename,'r') as f:
        data = f.readlines()
        for d in data:
            if d[-1] == '\n':
                d = d[:-1]
                
                tmp = d.split(',')
                name = tmp[0]
                sex = tmp[1]
                birth = tmp[2]
                
                if sex == 'F':
                    ret = nameF
                else:
                    ret = nameM
                    
                if name in ret:
                    ret[name] += int(birth)
                else:
                    ret[name] = int(birth)
                    
    retF = sorted(nameF.items(), key=lambda x:x[1],reverse=True)
    retM = sorted(nameM.items(), key=lambda x:x[1], reverse=True)

    for i, name in enumerate(retF):
        if i>9:
            break
        print('TOP_%d 여자아기이름:%s'%(i+1,name))
    for i, name in enumerate(retM):
        if i>9:
            break
        print('Top_%d 남자아기이름:%s'%(i+1,name))
y = input('출생년도 입력하세요(예:2001):')
getTop10BabyName(y)
```

```python
# age 변수를 10, 20, 30, 40, 50, 60이상 범주로 구분하여 age_cat 변수를 생성하세요.
tt.age
```

```python
def age_cat(age):
    cat=''
    if age <20: cat ='10'
    elif age<30: cat='20'
    elif age<40: cat='30'
    elif age<50: cat='40'
    elif age<60: cat='50'
    else: cat ='60'
    return cat
tt['age_cat'] = tt.age.apply(lambda x: age_cat(x))
tt.age_cat
tt[['age','age_cat']].head()
```
