```python
import urllib.request
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from urllib import parse
from bs4 import BeautifulSoup as bs

from datetime import datetime
today = datetime.now().strftime('%Y%m%d') # datetime 라이브러리를 통해 현재 날짜 today 생성

# 공공데이터포털에서 발급받은 서비스키
servicekey ='kLRBs4yEN7GJBLsGjh0ogEiI3aUVdSrNc8sPuiRxPG7sBSpDHWzAd2Uzy4RixuClUiP9dFhDhFgBIPVLQGymkA%3D%3D'
# 서비스URL
url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?'
params = {'serviceKey': parse.unquote(servicekey),  'startCreateDt':20200201,# 데이터 생성일 시작범위(선택)
          'endCreateDt':today} # 데이터 생성일 종료범위(선택)
                                        
# request = urllib.request.Request(url)
res = requests.get(url, params=params)
# 결과 출력
soup = bs(res.text, 'lxml')
print(soup)
```

```python
# 리스트 원소 확인
# 리스트의 원소는 <item> </item>으로 묶여있는 날짜별(statedt) 데이터

items = soup.find_all('item')
# print(items)
for x in items:
    print(x)
```

```python
# xml을 dictionary로 변환
response = {'seq': '고유번호',
            'statedt': '기준일',
            'statetime': '기준시간',
            'decidecnt': '확진자',
            'clearcnt': '격리해제',
            'examcnt': '검사진행',
            'deathcnt': '사망자',
            'carecnt': '치료중',
            'resutlnegcnt': '음성',
            'accexamcnt': '누적검사',
            'accexamcompcnt': '누적검사완료',
            'accdefrate': '누적확진률',
            'createdt': '등록일시',
            'updatedt': '수정'}

# xml을 dictionary로 바꿔 리스트 원소로 추가
lst = [] ## 이해안감
for y in items:
    l = {}
    for x in y:
        l[x.name] = x.text
    lst.append(l)
lst
```

```python
# list를 dataframe으로 만들기
import pandas as pd
df = pd.DataFrame(lst)
df.columns = df.columns.map(response)
df
```

```python
data = df.copy()
data.head()
```

```python
data.info()
```

```python
# data type 변경을 위해 "수정"컬럼의 날짜 및 시간정보에 포함된 문자열 'null'을 ''으로 수정
data['수정']
```

```python
data['수정'] = data['수정'].str.replace('null','')
data.수정 # null값 없어진거 확인
```

```python
# data type 확인 및 변경
for col in data.columns:
    if col in  ['등록일시','기준일','수정']: # 날짜: datetime
        data[col] = data[col].apply(pd.to_datetime) # to_datetime으로 변경
    elif col == '기준시간': # 시간: object
        data[col] = data[col].astype('O')
    else: # 나머지 columns: 수치형
        data[col] = data[col].apply(pd.to_numeric) #  문자열 칼럼의 숫자형 변환
data.info()
```

```python
# 2020-02-01 ~ 2022-02-13 today까지의 데이터 확인
data
```

```python
# 중복값 처리
# 보건복지부 코로나19 감염 현황 API에는 수정기록도 함께 있기 때문에 
# 기준일이 동일한 데이터가 존재하여 수정된 데이터만 남기고 이전 데이터는 삭제
# duplicated를 통해 중복값 확인
data = data.drop(index=data[data.기준일.duplicated(keep='first')].index).reset_index(drop=True)
data[data.기준일.duplicated(keep=False)]
data
```

```python
# 기준일 기준 내림차순 정렬
data = data.sort_values('기준일', ascending=False).reset_index(drop=True)
data
```

```python
# 일일현황 데이터 추가
# diff(-1) 끝에서부터 행 단위 차이 계산
daily = data.loc[:,['확진자','사망자','누적검사']].diff(-1)
daily.columns = ['일확진자','일사망자','일검사완료']
daily
```

```python
# 일일현황 데이터 추가
# diff(-1) 끝에서부터 행 단위 차이 계산
daily = data.loc[:,['확진자','사망자','누적검사']].diff(-1)
daily.columns = ['일확진자','일사망자','일검사완료']
daily
```

```python
# concat으로 data, daily 데이터프레임 결합
daily_data = pd.concat([data, daily], axis=1)
daily_data
```

```python
# DataFrame.loc[row label, column label]
ndf = daily_data.loc[:,['기준일','일확진자','일사망자','일검사완료']].dropna()
ndf.loc[: ,'일확진자':] = ndf.loc[: ,'일확진자':].astype(int)
ndf
```
