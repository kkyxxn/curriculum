```python
import urllib.request
import requests
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from urllib import parse
from bs4 import BeautifulSoup as bs

from datetime import datetime
today = datetime.now().strftime('%Y%m%d') # datetime     today 

#   
servicekey ='kLRBs4yEN7GJBLsGjh0ogEiI3aUVdSrNc8sPuiRxPG7sBSpDHWzAd2Uzy4RixuClUiP9dFhDhFgBIPVLQGymkA%3D%3D'
# URL
url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19InfStateJson?'
params = {'serviceKey': parse.unquote(servicekey),  'startCreateDt':20200201,#   ()
          'endCreateDt':today} #   ()
                                        
# request = urllib.request.Request(url)
res = requests.get(url, params=params)
#  
soup = bs(res.text, 'lxml')
print(soup)
```

```python
#   
#   <item> </item>  (statedt) 

items = soup.find_all('item')
# print(items)
for x in items:
    print(x)
```

```python
# xml dictionary 
response = {'seq': '',
            'statedt': '',
            'statetime': '',
            'decidecnt': '',
            'clearcnt': '',
            'examcnt': '',
            'deathcnt': '',
            'carecnt': '',
            'resutlnegcnt': '',
            'accexamcnt': '',
            'accexamcompcnt': '',
            'accdefrate': '',
            'createdt': '',
            'updatedt': ''}

# xml dictionary    
lst = [] ## 
for y in items:
    l = {}
    for x in y:
        l[x.name] = x.text
    lst.append(l)
lst
```

```python
# list dataframe 
import pandas as pd
df = pd.DataFrame(lst)
df.columns = df.columns.map(response)
df
```

```python
data = df.copy()
data.head()
```

```python
data.info()
```

```python
# data type   ""      'null' '' 
data['']
```

```python
data[''] = data[''].str.replace('null','')
data. # null  
```

```python
# data type   
for col in data.columns:
    if col in  ['','','']: # : datetime
        data[col] = data[col].apply(pd.to_datetime) # to_datetime 
    elif col == '': # : object
        data[col] = data[col].astype('O')
    else: #  columns: 
        data[col] = data[col].apply(pd.to_numeric) #     
data.info()
```

```python
# 2020-02-01 ~ 2022-02-13 today  
data
```

```python
#  
#  19   API     
#          
# duplicated   
data = data.drop(index=data[data..duplicated(keep='first')].index).reset_index(drop=True)
data[data..duplicated(keep=False)]
data
```

```python
#    
data = data.sort_values('', ascending=False).reset_index(drop=True)
data
```

```python
#   
# diff(-1)     
daily = data.loc[:,['','','']].diff(-1)
daily.columns = ['','','']
daily
```

```python
#   
# diff(-1)     
daily = data.loc[:,['','','']].diff(-1)
daily.columns = ['','','']
daily
```

```python
# concat data, daily  
daily_data = pd.concat([data, daily], axis=1)
daily_data
```

```python
# DataFrame.loc[row label, column label]
ndf = daily_data.loc[:,['','','','']].dropna()
ndf.loc[: ,'':] = ndf.loc[: ,'':].astype(int)
ndf
```
