```python
# [과제] t_df의 Age 컬럼을 이용하여 새로운 컬럼 Age_cat를 사용자 함수를 이용하여 출력하세요(7개 카테고리)
def get_category(age):
    cat = ''
    if age<10 : cat ='Child' # 10미만이면 child
    elif 10 <= age < 20: cat ='Teenage'
    elif 20 <= age < 30: cat = 'Twenty'
    elif 30 <= age < 40: cat = 'Thirty'
    elif 40 <= age < 50: cat = 'Forty'
    elif 10 <= age < 60: cat = 'Fifty'
    else : 
        cat ='Elderly'
    return cat

t_df['Age_cat']=t_df.Age.apply(lambda x :  get_category(x))
t_df[['Age', 'Age_cat']].head()
```

```python
# [과제] t_df에 대하여 결손값 처리, 불필요한 컬럼 삭제, 인코딩 등 을 수행하는 전처리 과정을 
# 하나의 사용자 함수(def transform_features(df))로 작성하여 분석용 데이터셋을 만드세요.
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv('./dataset/train.csv')
df

def transform_features(df):
    # 결손값 처리 -> fillna 
    # 결측값을 앞 방향(ffill) 혹은 뒷 방향(bfill)으로 채우기
    # fillna(method='ffill' or 'pad'), fillna(method='bfill' or 'backfill')
    
    df['Age'].fillna(df['Age'].mean(), inplace=True)
    df['Cabin'].interpolate(method='linear', inplace= True)
    df['Embarked'].fillna(method='ffill', inplace=True)
    

    # 불필요한 컬럼 삭제
    df.drop(['PassengerId','Name','Ticket'], axis=1, inplace= True)
    
    # 인코딩
    col=['Cabin','Sex','Embarked'] # 문자열을 인코딩 하기 위해 리스트로 넣어줌
    df['Cabin'] = df['Cabin'].str[:1] # cabin은 문자+숫자인데 숫자는 의미가 없다 판단하고 문자만 슬라이싱
    for f in col:
        le = LabelEncoder()
        le.fit(df[f])
        df[f] = le.transform(df[f])
        
    return df
transform_features(df)
```

```python
def transfrom_features(t):
    t1 = t.copy()
    t1['Family'] = t.SibSp + t.Parch
    
    # 사용하지 않을 열 삭제
    t1.drop(['PassengerId','Name','Ticket','Name_len','Age_cat','SibSp','Parch'], axis=1, inplace=True) 
    for i in t1.columns:
         # 결측값이 있는 것들
        if t1[i].isna().sum() != 0:                        
            # 결측값 개수가 데이터 개수의 70프로를 넘는 경우 그 열 안씀
            if t1[i].isna().sum().sum() > len(t_df) * 0.7:  
                t1.drop(i,axis=1,inplace=True)
            # 문자인 경우 최빈값으로 대체   
            elif t1[i].dtype == 'O':                        
                many = t1[i].value_counts().index[0]
                t1[i] = t1[i].fillna(many)
            # 숫자인 경우 평균값으로 대체    
            else:
                t1[i][t1[i].isna()] = t1[i].mean()         
    return pd.get_dummies(t1)
transfrom_features(t_df)
```

```python
# [과제] t_df Name 컬럼에서 Title(칭호)를 추출해서 이에 기반한 age 평균값을 이용하여 age null 처리를 한 후 개선된 부분을 설명하세요.
name_grade = t_df['Name'].apply(lambda x : x.split(", ",1)[1].split(".")[0])
name_grade = name_grade.unique().tolist()
print(name_grade)

# 호칭에 따라 사회적 지위(1910년대 기준)를 정의합니다.
grade_dict = {'A': ['Rev', 'Col', 'Major', 'Dr', 'Capt', 'Sir'], # 명예직
              'B': ['Ms', 'Mme', 'Mrs', 'Dona'],                 # 여성
              'C': ['Jonkheer', 'the Countess'],                 # 귀족이나 작위
              'D': ['Mr', 'Don'],                                # 남성
              'E': ['Master'],                                   # 어린남성
              'F': ['Miss', 'Mlle', 'Lady']}                     # 젊은 여성

# 이름 재정의 사용자 함수
def give_grade(x):
    grade = x.split(", ", 1)[1].split(".")[0]
    for key, value in grade_dict.items():
        for title in value:
            if grade == title:
                return key

# Name에 적용 및 확인
t_df['Name'] = t_df['Name'].apply(lambda x: give_grade(x))
print(t_df['Name'].value_counts())

# 사회적 지위에 따라 보통 나이가 비슷. 지위를 참고해 나이 결측치 대체
import warnings
warnings.filterwarnings('ignore')

for i in t_df.Name:
    t_df.Age[(t_df.Name == i)&(t_df.Age.isnull())] = t_df.Age[t_df.Name == i].mean()
t_df.info()
```
