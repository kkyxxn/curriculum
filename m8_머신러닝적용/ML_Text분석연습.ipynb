- 품사를 태깅한다는 것은 주어진 텍스트를 형태소 단위로 나누고, 나눠진 각 형태소를 그에 해당하는 품사와 함께 리스트화 하는 것을 의미한다.

Q. 'dataset/영화 기생충_review.txt' 문서를 불러와서 아래 사항을 수행하세요.

- 정규표현식을 이용해서 한글 단어외 모두 삭제 처리(사용자 함수 cleanText(readData) 작성하여 활용)
- 정제된 문서에서 명사, 동사, 형용사만 단어 리스트로 추출
- 단어 빈도 분석 후 빈도수에서 Top 10 단어를 막대 그래프로 시각화

```python
import pandas as pd
from konlpy.tag import Okt
import re

okt = Okt()

with open('dataset/영화 기생충_review.txt', 'r') as file:
    text = file.read()
    print(text)
```

```python
def cleanText(readData):
    text = re.sub('[0-9]+','', readData) # 숫자 한개 이상있으면 공백으로
    text = re.sub('[A-Za-z]+','', text) # 영어 대소문자 공백으로
    text = re.sub('[-=+,#/\?:^$.@*\"※~&%ㆍ·!』\\‘’|\(\)\[\]\<\>`\'…》]','',text) # 모든 특수문자 기호들 공백으로
    return text

text = cleanText(text)

# 명사, 동사, 형용사만 추출
morph = okt.pos(text) # 단어, 품사 같이 태깅
word_list = [] # 그걸 담을 그릇 세팅
for word, tag in morph: # word와 tag를 하나씩 뽑아줌
    if(tag=='Noun')|(tag=='Verb')|(tag=='Adjective'): # tag가 명사이거나 동사이거나 형용사이면
        word_list.append(word) # 단어들을 word_list에 뿌려줘라 
print(word_list)
```

```python
# 단어별 빈도
# collections.Counter(a) : a에서 요소들의 개수를 세어, 딕셔너리 형태로 반환합니다.  {문자 : 개수} 형태
from collections import Counter
count = Counter(word_list) # word_list에서 단어들을 count해줌
# collections.Counter(a).most_common(n)   : a의 요소를 세어, 최빈값 n개를 반환합니다. (리스트에 담긴 튜플형태로)
n_words = dict(count.most_common())
print(n_words)
```

```python
data = pd.Series(n_words)
df = pd.DataFrame(data, columns=['cnt'])
df10 = df.sort_values(by='cnt', ascending=False).head(10)
df10
```

```python
df10.reset_index(inplace=True)
df10.head()
```

### merge
- 기준열 이름이 같을 때
    - pd.merge(x, y, on = '기준열', how = '조인방식')
- 기준열 이름이 다를 때
    - pd.merge(x, y, left_on = '왼쪽 열', right_on = '오른쪽 열', how = '조인방식')<br><br>
    
##### 병합 방식
- left_on : 기준열 이름이 다를 때, 왼쪽 기준열
- right_on : 기준열 이름이 다를 때, 오른쪽 기준열<br><br>

```python
pd.merge(grade, fruit, on='Num', how='left') # 칼럼순서가 Num -> grade -> fruit
```

```python
# right Join -> 오른쪽 데이터프레임을 기준으로 조인한다. 왼쪽 데이터프레임에 없는 값은 NaN으로 나타난다.
pd.merge(fruit, grade, on = 'Num', how = 'right')
```

```python
# Inner Join 교집합을 의미한다. 양쪽에 공통으로 있는 값만 나타난다. 
pd.merge(fruit, grade, on = 'Num', how='inner')
```

Konlpy를 이용해 text의 형태소 분석하여 텍스트 전처리

```python
# 파일 안에 텍스트들을 konlpy를 이용해 명사, 형용사만 추출
from konlpy.tag import Okt
from collections import Counter

# Okt 함수를 이용해 형태소 분석
okt = Okt()
# 리스트 초기화 및 선언
line = []
# blog_text에서 각 품사를 태깅하고 리스트화
line = okt.pos(blog_text)

n_adj = []

# 품사를 태깅한 리스트에서 명사 또는 형용사 단어만 n_adj에 넣어주기
for word, tag in line:
    if tag in ['Noun','Adjective']:
        n_adj.append(word)
print(n_adj)

# 출력 결과를 보면 '아니었네요.', '끝'등과 같은 워드클라우드를 만들 때 부적절한 단어가 포함되어있다.
# 이와 같은 단어를 불용어라하는데 이러한 불필요한 단어는 삭제시킨다.
```

불용어(stopwords) 리스트에서 삭제

```python
#제외할 단어 추가
stop_words = "아니었네요. 있어서 끝 " #추가할 때 띄어쓰기로 추가해주기
stop_words = set(stop_words.split(' '))
# 불용어를 제외한 단어만 남기기
n_adj = [word for word in n_adj if not word in stop_words]
```

```python
# 가장 많이 나온 단어 100개 저장
counts = Counter(n_adj)
tags = counts.most_common(100)
```

```python
print(tags)
```

워드클라우드 만들기
- 데이터프레임을 워드클라우드로 만들려면 사전(dictionary) 형태로 변환을 해줘야 한다.
- 이미 사전형이라면 굳이 변환을 해주지 않아도 된다. collections  모듈의 Counter 함수를 사용하면 자동으로 사전형으로 변환이 된다.

```python
# 마스크라는 기능을 이용하여 원하는 모양의 워드클라우드를 만들 수 있다.
# 이미지의 경계선이 뚜렷한 png 파일이 필요하다.
import numpy as np
from PIL import Image

# 이미지 추가(워드크라우드 모양 설정)
mask = Image.new("RGBA",(2555,2275), (255,255,255)) #(2555,2575)는 사진 크기, (255,255,255)는 색을의미
image = Image.open('C:\\cakd5\\workspace\\m8_머신러닝적용\\dataset\\heart.png')
x,y = image.size
mask.paste(image,(0,0,x,y),image)
mask = np.array(mask)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
%matplotlib inline

wc = WordCloud(font_path ='malgun', #폰트
               background_color = 'black', #배경색
               width = 800, height = 600, 
               # mask=mask
               max_font_size=250,  #가장 많이 나온 단어 fontsize 설정
               scale=2.0)

gen = wc.generate_from_frequencies(dict(tags))
# .generate(' '.join(train_data_html_out))

plt.figure(figsize = (15, 10)) # 각각 가로 세로 인치 
plt.axis("off") # 축에 표시되는 눈금을 제거하는 옵션
plt.imshow(gen) # 이미지가 표시되도록 하는 옵션

# 워드클라우드 사진으로 저장
# cloud.to_file('상암맛집_wc.png')
plt.savefig('상암맛집.png', bbox_inches='tight')
plt.show()
```

### 텍스트 ML process
- 텍스트 처리해주기 위해서 텍스트 정규화 > 전처리하고 벡터화
- 피처 벡터화 
    - CountVectorizer  빈도수만 적용
    - TFi-IDF 전체 문서에서 조사 등과 같이 자주 쓰이지만 의미없는 단어에 패널티 적용하여 CountVectorizer를 보완해줌
- 댓글에 대해 부정, 긍정에 대한 감성분석
- 문서를 군집화 > k-means를 사용하기 위해 피처벡터화 먼저함
- 문서 유사도 측정 Cosine Simularity 사용
