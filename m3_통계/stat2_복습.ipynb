### 확률의 수학적 정의와 의미
   - 표본공간과 확률표본  -> 표본은 모집단에서 샘플을 뽑는 것
      - 확률표본 또는 표본은 풀고자 하는 문제에서 발생할 수 있는 하나의 현상 혹은 선택될 수 있는 하나의 경우를 말한다.<br><br>
      - 표본공간은 가능한 모든 표본의 집합을 말하며 고려하는 범위에서 어떤 표본(경우, 현상)이 가능하고 어떤 표본이 가능하지 않은가를 정의하는 작업
           * ex) 주사위 6개 나올 수 있는 모수는 ? 123456  
           * 표본은 ? 123456 모두 가능
           * 표본의 집합 표본의 공간은 ? 123456
           * 7은 표본공간에 없으므로 안됨<br><br>
           
     - 사건은 표본공간의 부분집합, 즉 전체 표본공간 중에서 관심을 가지는 일부 표본의 집합
          * 123456이 있는데 3이 나오면 그게 사건<br><br>
     - 표본과 사건의 차이는 관심유무 3이란 숫자는 표본도 될 수 있고 사건도 될 수 있지만 관심을 가지므로 사건이 됨
      
### 확률
   - 확률이란 사건(부분집합)을 입력하면 숫자가(확률값) 출력되는 함수 
       * 3이 나올 확률 -> 1/6 -> 이것을 확률값이라 칭함<br><br>
   - 모든 각각의 사건(부분집합)에 어떤 숫자를 할당하는 함수<br><br>
   - 확률은 표본이 아닌 사건을 입력으로 가지는 함수

### 확률의 의미
   - 빈도주의 관점 : 반복적으로 선택된 표본이 사건 A의 원소가 될 경향 --> 동전을 던져 앞면이 나오는 사건의 확률값은 0.5<br><br>
   - 베이지안 관점 : 선택된 표본이 특정한 사건에 속한다는 가설, 명제 혹은 주장의 신뢰도로 반복이란 개념은 사용되지 않음 --> 새가 날 수 있는 가능성은 95%, '앞면이 나왔다'는 주장의 신뢰도가 0.5
       - 베이지안 확률론에서 사건이 일어났다 혹은 발생했다하는 말은 그 사건의 원소 중에 정말로 선택된 표본이 있다는 사실을 알게되었다는 것을 말한다.

### 확률의 성질
   - 공집합의 확률 : 공집합인 사건의 합은 0이다. --> P(0) = 0<br><br>
   - 여집합의 확률 : 어떤 사건의 여집합인 사건의 확률은 1 - 원래 사건의 확률과 같다.<br><br>
   - 포함-배제의 원리 : 두 사건의 합집합의 확률은 각 사건의 확률의 합에서 두 사건의 교집합의 확률을 뺀 것과 같다.<br><br>
   - 전체 확률의 법칙 : 생일이 1월인 남자라는 사건의 확률과 생일이 2월인 남자라는 사건의 확률, 이외에도 각각의 월을 생일로 가지는 남자라는 사건들의 확률을 모두 합하면 남자라는 사건의 확률이 된다. 각각의 월에 있는 남자의 생일이 서로 공통 원소가 없음

### 확률분포함수 -> 확률변수와 확률간의 분포
   - 확률이 어디에 어느 정도 분포되어 있는지를 수학적으로 명시하고 전달하는 도구<br><br>
   - 만약 표본의 개수가 무한하다면 현실적으로 모든 사건을 하나하나 기술하는 것은 불가능하며 확률분포함수를 이용하여 이 문제를 해결<br><br>
       - 확률질량함수 : 유한 개의 사건이 존재하는 경우 각 단수사건에 대한 확률만 정의하는 함수
           * ex)  동전 던지기 앞뒤 딱 나오는것, 이산형 데이터<br><br>
       - 누적분포함수 : 시작점을 모두 똑같이 음의 무한대로 통일한 특수한 구간을사용하는 사건의 확률분포를 묘사하는 함수<br><br>
       - 확률밀도함수 : 누적분포함수를 미분하여 구한 도함수를 의미하며 특정한 구간의 확률이 다른 구간에 비해 상대적으로 얼마나 높은가를 나타내는 것이며 그 값 자체가 확률은 아니다.
           * 누적분포, 확률밀도는 연속형 데이터

        - 동전 던지기의 확률변수는? 앞 뒤 / 주사위의 확률변수는? 1 2 3 4 5 6

### [모집단 분포와 표본분포]
 - 모집단 : 어떤 통계적 실험의 대상이 되는 모든 대상물. ex) 대한민국 모든 남성의 키 데이터, 2018 수능점수 데이터<br><br>
 - 모집단 분포 : 모집단은 데이터 집단이므로 어떠한 분포를 이루며 이 데이터가 이루는 확률 분포가 모집단 분포<br><br>
 - 모수 : 모집단의 특성을 나타내는 수치로 모평균, 모분산, 모표준편차, 모비율, 모상관관계 등<br><br>
 - 표본 : 모집단의 모수(특성)을 알기 위하여 표본을 추출해서 모집단의 모수를 추정<br><br>
 - 확률표본 : 모집단을 구성하는 모든 구성원이 표본으로 뽑힐 확률을 동일하게 하여 추출한 표본. Random이란 무작위라는 의미로 어떠한 의도를 가지지 않는다는 것을 의미하며 모든 구성원이 뽑힐 확률이 동일하다는 개념<br><br>
 - 통계량 : 확률표본의 특성을 설명하는 값으로 표본평균, 표본분산, 표본표준편차, 표본비율, 표본상관관계 등 모수는 모집단이 변하지 않기 때문에 그 값이 변하지 않으나 통계량은 표본을 어떻게 추출하느냐에 따라서 그 값이 다르게 나타난다.<br><br>
 - 확률변수 : 무작위 실험을 했을 때 특정 확률로 발생하는 각각의 결과를 수치값으로 표현한 변수. 동전의 앞과 뒤<br><br>
 - 확률분포 : 확률변수 X의 모든 값과 그에 대응하는 확률값들의 분포.<br><br>
 - 통계량이 확률변수라는 의미 : 확률변수의 확률분포가 수학적으로 설명가능한 분포(예로 정규분포)를 따른다고 하면 특정 확률변수와 확률함수를 알고 있다면 특정 사건이 일어날 확률을 계산할 수 있으며 -> why? 분포를 알고 있으므로<br><br>
 더 나아가 모수와 통계량의 관계를 알고있다면 그것을 통해 모수에 대한 통계적인 추정을 할 수 있다. 즉 통계량의 확률 함수 또는 표본분포를 안다고 가정하면 이를 이용해 모수를 추정할 수 있다.

### 따라서 확률분포냐 정규분포냐 이항분포냐 포아성분포냐 지수분포냐 이런것들이 base가 된다.
1.표준 정규분포는 평균이 0 편차가 1인 분포임 우리가 어떤 확률변수 x1 x2를 알면 정규분포라는 함수에 대응해서 y값을 알 수 있다. <br><br>
2. 그러면 모수가 이런 정규분포를 따르고 있다고 가정하고 우리가 모수와 통계량의 관계를 알고 있다면 그것을 통해 모수에 대한 통계적인 추정을 할 수 있다.<br><br>
3. 즉 모수가 정규분포를 따르고 샘플이 많아지면 표본의 평균은 모집단의 평균에 근사하다.

### [표본통계량 성질]
 - 샘플사이즈가 크면 표본평균은 모평균에 가까워진다.<br><br>
 - 샘플사이즈가 커질수록 표본평균의 표준편차가 작아진다. 샘플사이즈를 크게하면 흩어짐이 적은 신뢰할 수 있는 표본평균을 얻을 수 있다.<br><br>
 - 표준오차는 이론상 표본평균의 표준편차 크기인데 샘플사이즈가 커지면 표준오차는 작아짐<br><br>
 - 모분산은 0.64, 표본분산의 평균값은 0.575로 분산이 과소평가 되어짐 -> 불편분산을 사용 편향을 제거(ddof=1)<br><br>
 - 샘플사이즈가 커질수록 불편분산은 모분산에 가까워짐
     - 불편분산은 과소평가되는것을 보정하기위함
     - 그렇다고 모분산과 똑같진 않다 샘플사이즈가 크면 클수록 모분산에 근접할뿐<br><br>
 - 불편성 : 추정량의 기댓값이 진짜 모수가 되는 특성. 편향이 없는 추정량<br><br>
 - 일치성 : 샘플사이즈가 커지면 추정량이 진짜 모수에 가까워지는 특성<br><br>
 - 표본평균과 불편분산은 둘다 불편성을 가지고 있다고 할 수 있다.<br><br>
 - 샘플사이즈를 크게하면 표본평균은 모평균에 가까워지고 불편분산은 모분산에 가까워지므로 일치성을 가짐<br><br>
 - 큰수의 법칙 : 표본의 크기가 커지면 표본평균이 모평균에 가까워지는 방법을 표현한 법칙<br><br>
 - 중심극한정리 : 모집단분포가 무엇이든지간에 샘플사이즈가 커지면 표본 평균들이 이루는 분포가 모집단의 평균 𝜇,표준편차𝜎/𝑠𝑞𝑟𝑡(𝑛)인 정규분포에 가까워진다는 정리 -> 어느 특정 분포만이 아니란 뜻, 다른 여러 분포들에도 표본의 표본평균들이 이루는 분포는 샘플사이즈가 커지면 정규분포에 근사함

```python
# 모집단과 표본
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('./data/ch4_scores400.csv')
df.shape
scores = np.array(df['score'])
scores[:10] # 1차원 10개 뽑기
```

```python
# 확률분포(균형이 안맞는 주사위를 가정)
dice = [1,2,3,4,5,6] # 주사위 눈의 개수
prob = [1/21,2/21,3/21,4/21,5/21,6/21] # 뽑힐 확률
```

```python
# 0~2까지의 숫자를 10번 무작위 추출
np.random.choice(3,10)
```

```python
# 주사위 눈 6개 각각의 비율대로 무작위 추출
np.random.choice(dice,p=prob) 
```

```python
np.random.seed(1)
num_trial = 100 # 주사위 100번 시도
sample = np.random.choice(dice, num_trial, p=prob)
sample
```

```python
# 도수분포표로 각 눈이 나온 횟수와 비율을 파악
freq, _ = np.histogram(sample,bins=6, range=(1,7))
print(freq,'\n')
rdf = pd.DataFrame({'frequency':freq,
                   'relative frequency':freq/num_trial},
                  index = pd.Index(np.arange(1,7), name='dice'))
rdf
```

```python
# 샘플사이즈를 순차적으로 늘렸을 때 표본평균이 모평균에 가까워짐을 시각화하여 입증하세요.
# 큰(대수)의 법칙 시각화
from scipy.stats import norm
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

result = pd.DataFrame(columns=['sample_num','mean']) # 컬럼명 생성

# result
for i in range(100,10001):
    pop = norm(loc = 4,scale = 0.8) # 기댓값 4 표준편차 0.8 생성하고 pop에 담아
    
   # 무작위 표본 생성 -> rvs(random value sampling)
    choice = pop.rvs(size = i) # 무작위 표본 생성 사이즈는 i로 -> i개의 난수 생성해서 choice에 담아
    
    mean = choice.mean() # 그 난수의 평균을 mean에 담고
    # 데이터프레임의 데이터값으로 그 난수와 난수의 평균을 뿌려줘라 
    # ignore_index = False, False값은 기존 index 유지, True값은 기존 index 무시
    
    # append함수로 뿌리기 -> 기댓값 4 표준편차 0.8인 난수(i)와 그 난수의 평균값 
    result = result.append(pd.DataFrame([[i, mean]], columns=['sample_num', 'mean']), ignore_index=True)
    
fig = plt.figure(figsize = (35, 20))
sns.lineplot('sample_num','mean',data = result) # lineplot(x축, y축, 보여줄 데이터)
plt.axis([100,10000,3.8,4.2]) # x축 시작 끝, y축 시작 끝 설정
```

```python
# 샘플사이즈가 커질수록 표본평균의 표준편차가 작아짐
# why? 샘플 사이즈가 커질수록 표본평균은 모평균에 근사하게 됨으로 당연히 표준편차도 줄어들게 된다
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
from scipy import stats

size_array  = np.arange(10,10000)
print('초기의 size_array크기 : ',(len(size_array))) # 9990

population = stats.norm(loc = 4, scale = 0.8) # 기댓값4 표준편차 0.8인 정규분포
population.rvs(size = 10) # 무작위 표본 생성 -> rvs(random value sampling) 10개 추출

# 샘플사이즈를 변화시키면서 표본평균을 도출하는 시뮬레이션
np.random.seed(1)
sample_mean_array_size = np.zeros(len(size_array)) # zeros 배열 생성 -> sample_mean_array_size 담아
aaa = np.zeros(len(size_array)) # 똑같이 aaa에도 9990개 담아
print('for문 돌리기 전 aaa 크기 : ',(len(size_array)))

for i in range(0,len(size_array)): # 0부터 9989 -> 총 9990번 돌리면서 i에 할당
    
    # 평균이 4 표준편차가 0.8일때 무작위 랜덤 표본 난수 9990개 생성 -> sample 담고
    sample = population.rvs(size = size_array[i])  
    
    # 평균이 4 표준편차가 0.8일때의 무작위 랜덤으로 생성된 표본들의 평균값들을 sample_mean_array_size 담아
    sample_mean_array_size[i] = np.mean(sample)  
    # aaa[i]는 표준편차 / 
    aaa[i] = population.std() / (len(sample) ** 0.5) # x ** 0.5 -> x의 루트 씌운것 = math.sqrt(x)
sample_mean_array_size[:10] # 10개 출력
# print(len(sample))

# 샘플사이즈가 커질수록 표본평균이 모평균에 가까워짐
import matplotlib.pyplot as plt
plt.plot(size_array, aaa, color='black')
plt.xlabel('sample size')
plt.ylabel('sample std')
```
